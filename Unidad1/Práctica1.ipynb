```python
#libreria para generar numeros aleatorios
import random
import statistics
import matplotlib.pyplot as plt
fabrica1=[]
for i in range(100):
   numero_aleatorio=round(random.uniform(50,100),2)
   fabrica1.append(numero_aleatorio)
print(fabrica1)
plt.hist(fabrica1)
plt.title("Produccción diaria de la fabrica 1")
media=statistics.mean(fabrica1)
mediana=statistics.median(fabrica1)
moda=statistics.mode(fabrica1)
print("la media es =>", media)
print("la mediana es =>", mediana)
print("la moda es =>", moda)
plt.hist(fabrica1,bins=10,alpha=0.3,color="b",edgecolor="black")
plt.axvline(media,color="r",linestyle="dashed",label="media")
plt.axvline(mediana,color="g",linestyle="dashed",label="mediana")
plt.axvline(moda,color="y",linestyle="dashed",label="moda")
plt.xlabel("frecuencia")
plt.ylabel("producción diaria")
plt.legend()
plt.show()
```

```python
import random
import statistics
import matplotlib.pyplot as plt
fabrica2=[]
for i in range(100):
    numero_aleatorio=round(random.uniform(40,95),2)
    fabrica2.append(numero_aleatorio)
plt.hist(fabrica2)
plt.title("Produccción diaria de la fabrica 2")
media2=statistics.mean(fabrica2)
mediana=statistics.median(fabrica2)
moda=statistics.mode(fabrica2)
print("la media es =>", media2)
print("la mediana es =>", mediana)
print("la moda es =>", moda)
#marcar en grafica
plt.title("Produccción diaria de la fabrica 2")
media2=statistics.mean(fabrica2)
mediana2=statistics.median(fabrica2)
moda2=statistics.mode(fabrica2)
print("la media es =>", media2)
print("la mediana es =>", mediana2)
print("la moda es =>", moda2)
plt.hist(fabrica2,bins=10,alpha=0.3,color="b",edgecolor="black")
plt.axvline(media2,color="r",linestyle="dashed",label="media")
plt.axvline(mediana2,color="g",linestyle="dashed",label="mediana")
plt.axvline(moda2,color="y",linestyle="dashed",label="moda")
plt.xlabel("frecuencia")
plt.ylabel("producción diaria")
plt.legend()
plt.show()
```

```python
#genererar 2 subplot de 1 renglon por 2 columnas para desplegar los 2 histograms; una al ladao de la otra.
import matplotlib.pyplot as plt
fig,(ax1,ax2)=plt.subplots(nrows=1, ncols=2, figsize=(10,5))
ax1.hist(fabrica1)
ax1.set_title("Produccción diaria de la fabrica 1")
ax2.hist(fabrica2)
ax2.set_title("Produccción diaria de la fabrica 2")
plt.show()
```
